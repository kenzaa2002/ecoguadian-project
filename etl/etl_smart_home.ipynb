```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df_home = pd.read_csv("C:/Users/lenovo/Desktop/EcoGuardian/data/energy_dataset.csv")
```

```python
print(f"Dataset shape: {df_home.shape}")
print(f"\nColumns: {list(df_home.columns)}")
print(f"\nFirst few rows:")
df_home.head()
```

```python
# Data exploration and preprocessing
print("Dataset Info:")
df_home.info()

print("\nMissing values:")
print(df_home.isnull().sum())

print("\nStatistical summary:")
df_home.describe()
```

```python
df_home["Date"] = pd.to_datetime(df_home["Date"], errors="coerce") 
df_home["Time"] = pd.to_datetime(df_home["Time"], errors="coerce").dt.time
```

```python
df_home
```

```python
le = LabelEncoder()
df_home["season_encoded"] = le.fit_transform(df_home["Season"])
```

```python
label_encoder = LabelEncoder()
df_home["Appliance_Type_Label"] = label_encoder.fit_transform(df_home["Appliance Type"])
df_home = df_home.drop(columns=["Appliance Type"])
```

```python
df_home = df_home.drop(columns=["Season"])
df_home
```

```python
df_home["Energy Consumption (kWh)"] = pd.to_numeric(df_home["Energy Consumption (kWh)"], errors="coerce") 
df_home["Outdoor Temperature (°C)"] = pd.to_numeric(df_home["Outdoor Temperature (°C)"], errors="coerce") 
df_home["Household Size"] = pd.to_numeric(df_home["Household Size"], errors="coerce")
```

```python
df_home = df_home.ffill().bfill()
```

```python
#features temporelles
df_home["weekday"] = df_home["Date"].dt.weekday
```

```python
df_home = df_home.drop(columns=["hour"])
```

```python
df_home.to_csv("C:/Users/lenovo/Desktop/EcoGuardian/data/smart_home_clean.csv", index=False) 
print(" Dataset Smart Home nettoyé et enrichi sauvegardé : smart_home_clean.csv")
```
